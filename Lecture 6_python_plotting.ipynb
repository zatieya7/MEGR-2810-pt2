### Lambda Functions in Python
A `lambda` function in Python is an anonymous, single-expression function. It is defined using the `lambda` keyword and allows for the creation of small, inline functions without needing to use the def keyword or assign a name. `lambda` functions are particularly useful in situations where a small, one-off function is needed as an argument to a higher-order function (a function that takes other functions as arguments). We will discuss more in future lectures. 

```python
square = lambda x: x**2
print(square(5))  # Output: 25
```

```python
# Sort a dictionary using only the built-in sorted() function.
# Show the original dictionary, the sorted keys, and items sorted by value.

items = {'Milk': 4.5, 'Bread': 3.2, 'Eggs': 6.0, 'Cheese': 8.5, 'Fruits': 12.0, 'Vegetables': 10.0, 'Snacks': 9.0, 'Beverages': 7.5}

# Print the original dictionary
print("Original items (dict):", items)

# sorted() on a dict returns a sorted list of keys by default
sorted_keys = sorted(items)
print("Sorted keys:", sorted_keys)

# To sort by values, use sorted() on items() with a key function (still using only sorted())
sorted_by_value = sorted(items.items(), key=lambda kv: kv[1]) # Always acending. Need to use lambda function. 
sorted_by_value_decending = sorted(items.items(), key= lambda kv:kv[1], reverse = True) # Need to add reverse = True
print("Items sorted by value (ascending):", sorted_by_value)
print("Items sorted by value (decending):", sorted_by_value_decending)
```

```python

items = {
    'Milk': 4.5,
    'Bread': 3.2,
    'Eggs': 6.0,
    'Cheese': 8.5,
    'Fruits': 12.0,
    'Vegetables': 10.0,
    'Snacks': 9.0,
    'Beverages': 7.5
}


print("Original items (dict):", items)


sorted_keys = sorted(items)
print("Sorted keys (alphabetically):", sorted_keys)

sorted_by_value = sorted(items.items(), key=lambda kv: kv[1])
print("Items sorted by price (ascending):", sorted_by_value)


sorted_by_value_descending = sorted(items.items(), key=lambda kv: kv[1], reverse=True)
print("Items sorted by price (descending):", sorted_by_value_descending)
```

## Use NumPy to creating arrays: `array`, `linspace`, `arange`, helpers

```python
# Please note that I intentionally did not add comments. 
# Add comments to code in this notebook.

import numpy as np

# 1D arrays 
# Create a 1D NumPy array of floats
a = np.array([1.2, 3.5, 5.4, 7.1], dtype=float)
# Create a 1D array of evenly spaced numbers from 0.0 to 5.0 (inclusive)
x_lin = np.linspace(0.0, 5.0, 11)     # inclusive ends
# Create a 1D array using arange(start, stop, step)
# This goes from 0.0 up to (but not including) 5.0, increasing by 0.5

x_rng = np.arange(0.0, 5.0, 0.5)      # half-open [start, end)
## Print the arrays to show the results
print('a      =', a)
print('lin    =', x_lin)
print('arange =', x_rng)

# 2D helpers
# Create a 2x3 array of all zeros
# zeros((rows, columns))

Z = np.zeros((2,3))
# # Create a 3x2 array of all ones
# ones((rows, columns))
O = np.ones((3,2))
# Create a 3x3 identity matrix (ones on diagonal, zeros elsewhere)
# eye(n) creates an n×n identity matrix
I = np.eye(3)
# Print each array to visualize their contents
print('zeros:') 
print(Z)
print('ones:')  
print(O)
print('eye:')
print(I)
```

```python
# Try comment out the lines of code below and run it to see the plot
# So you can see how the plot changes
# Add comments to explain what each line does
import matplotlib.pyplot as plt
import numpy as np
# Create an array of 100 evenly spaced points between -5 and 5
x = np.linspace(-5, 5, 100) # Change the range and number of points to see what happens
# Define the mathematical relationship (y = x²)
y = x**2
# Plot x and y on a graph
plt.plot(x, y)
# Add a title and labels for clarity
plt.title("Parabola: $y=x^2$")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
# Add a background grid for better visibility
plt.grid()
# plt.xlim(0, 5)  # Uncomment to limit x-axis from 0 to 5
# plt.ylim(0, 25) # Uncomment to limit y-axis from 0 to 25
# Add a horizontal line at y=0 for reference
plt.axhline(0, color='red',linewidth=0.5, ls='--')
#shows the plot
plt.show()
```

## Multiple 2D Plot Types

```python
# Try comment out the lines of code below and run it to see the plot
# So you can see how the plot changes
# Add comments to explain what each line does
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0, 11)
# y= x**2
y = np.power(x, 2)

plt.figure(figsize=(14, 8))
plt.subplot(2, 3, 1); plt.plot(x, y); plt.title('Plot'); plt.grid()
plt.subplot(2, 3, 2); plt.scatter(x, y); plt.title('Scatter'); plt.grid()
plt.subplot(2, 3, 3); plt.bar(x, y); plt.title('Bar'); plt.grid()
plt.subplot(2, 3, 4); plt.loglog(x, y); plt.title('Loglog'); plt.grid()
plt.subplot(2, 3, 5); plt.semilogx(x, y); plt.title('Semilogx'); plt.grid()
plt.subplot(2, 3, 6); plt.semilogy(x, y); plt.title('Semilogy'); plt.grid()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return np.sin(x) + np.cos(2*x)

# Define x values
x = np.linspace(-10, 10, 100)  # 100 points for clarity
y = f(x)

# Create subplots: 3 rows, 1 column
fig, axs = plt.subplots(3, 1, figsize=(8, 12))  # 3 plots vertically

# Line plot
axs[0].plot(x, y, color='blue', label='Line plot')
axs[0].set_title("Line Plot")
axs[0].set_xlabel("x")
axs[0].set_ylabel("y")
axs[0].grid(True)
axs[0].legend()

# Scatter plot
axs[1].scatter(x, y, color='red', label='Scatter plot')
axs[1].set_title("Scatter Plot")
axs[1].set_xlabel("x")
axs[1].set_ylabel("y")
axs[1].grid(True)
axs[1].legend()

# Bar plot
axs[2].bar(x, y, color='green', label='Bar plot', width=0.2)
axs[2].set_title("Bar Plot")
axs[2].set_xlabel("x")
axs[2].set_ylabel("y")
axs[2].grid(True)
axs[2].legend()

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

Problems like this will appear in exams. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for SHM
A = 5          # amplitude
omega = 2      # angular frequency
phi = 0        # phase in radians
t_max = 10     # total time
dt = 0.01      # time step

# Time array
t = np.arange(0, t_max, dt)

# SHM position function
x = A * np.cos(omega * t + phi)

# Optional: compute velocity and acceleration
v = -A * omega * np.sin(omega * t + phi)
a = -A * omega**2 * np.cos(omega * t + phi)

# Plot position vs. time
plt.figure(figsize=(10, 6))
plt.plot(t, x, label='Position x(t)', color='blue')
plt.plot(t, v, label='Velocity v(t)', color='red', linestyle='--')
plt.plot(t, a, label='Acceleration a(t)', color='green', linestyle=':')
plt.title("Simple Harmonic Motion Tracker")
plt.xlabel("Time (s)")
plt.ylabel("x, v, a")
plt.grid(True)
plt.legend()
plt.show()
```

## 3D Plotting Examples
This section is optional. Proceed if you are interested.

```python
# Create a 3D parametric plot for x=cos⁡(t),y=sin⁡(t),z=t
import numpy as np
import matplotlib.pyplot as plt


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
t = np.linspace(0, 10*np.pi, 500)
x = np.sin(t)
y = np.cos(t)
ax.plot3D(x, y, t)
ax.set_title("3D Parametric Plot")
plt.show()
```

```python
# 3D surface plot of Z=exp(−X^2−Y^2)- Gaussian Function
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create grid
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Compute Z
Z = np.exp(-X**2 - Y**2) # Z=exp(−X^2−Y^2)- Gaussian Function

# Plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis')

# Labels and title
ax.set_title('z = exp(-x² - y²))')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Colorbar
fig.colorbar(surf, shrink=0.5, aspect=10)

plt.show()
```
